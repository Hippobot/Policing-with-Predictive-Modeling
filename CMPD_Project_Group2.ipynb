# Project Details
---


**Research Topic**:

    Vehicle Theft in Charlotte, NC
    
**Problem Statement**:

    Vehicle Theft has shown consistent prevalence in urban areas, causing significant financial and emotional distress to citizens. Especially concerning is the occurrence of this type of crimes within private residences. Understanding the patterns, hotspots, and potential socio-economic correlations can aid in devising effective prevention and response strategies to aid in the contributions of predictive policing.

**Primary Research Question**:

    How are vehicle-theft related crimes in private residences spatially distributed in Charlotte, NC, from 2011 to 2016? Furthermore, how do these spatial patterns correlate with socio-economic factors, vehicle attributes, and victim characteristics?
 

**Sub-questions**:

    Are there specific residential areas (census tracts) that are consistently more vulnerable to these crimes?
    Is there a discernible correlation between the poverty rate of a census tract and the frequency of vehicle crimes?   Spoiler: No
    How have patterns of vehicle crimes in private residences changed over the years from 2011 to 2016?

```python
# List of columns to drop (Manually chosen, omits redundant features)
columns_to_drop = ['OBJECTID', 'TARGET_FID','Direction', 'Street_Type', 'Suffix', 'Apt_No',
                   'City', 'State', 'Zipcode', 'Reported_Date','Location_Desc', 
                   'Incident_From_Time', 'Incident_To_Date','Incident_To_Time','Clearance_Status', 
                   'Clearance_Date', 'Reporting_Agency', 'Follow_up_Section', 'OBJECTID.1']

vc_data = vc_data.drop(columns=columns_to_drop) # Drop columns from the main dataset
```

    Correcting Empty Space

```python
# Columns that appear to have empty/white space/NaN values
columns_with_empty = ['Vehicle_Year', 'Vehicle_Make', 'Vehicle_Model',
                      'Vehicle_Color', 'Vehicle_Body', 'Victim_No', 'Victim_Age',
                      'Victim_Race', 'Victim_Gender', 'Victim_Ethnicity']

# Replace empty spaces with NaN
for column in columns_with_empty:
    vc_data[column] = vc_data[column].replace(r'^\s*$', np.nan, regex=True)

# Drop rows containing NaN values in the specified columns
vc_data = vc_data.dropna(subset=columns_with_empty)
```

    Normalize 'Place2' variable

```python
# Strip white space of values in Place2 column and assign it back to the dataframe
vc_data['Place2'] = vc_data['Place2'].str.strip()

# Manual Dummy Variable;     1: 'Private Residence'/'Apartment/Duplex Private Res'  0: All other values
vc_data['Place2'] = vc_data['Place2'].apply(lambda x: 1 if x in ('Private Residence', 'Apartment/Duplex Private Res') else 0)

# Visualize Normalization
vc_data.groupby("Place2").size().sort_values()
```

Creating a new season column from the Incident_From_Date column

```python
#Drop Null values from Incident_From_Date

vc_data['Incident_From_Date'] = pd.to_datetime(vc_data['Incident_From_Date'], errors='coerce')

# Drop rows where 'Incident_From_Date' is null
vc_data = vc_data.dropna(subset=['Incident_From_Date'])
```

```python
# Define a function to bin dates into seasons
def bin_season(month):
    if month in [12, 1, 2]:
        return 1  # Winter
    elif month in [3, 4, 5]:
        return 2  # Spring
    elif month in [6, 7, 8]:
        return 3  # Summer
    elif month in [9, 10, 11]:
        return 4  # Autumn

# Apply the modified binning function to create the new 'Seasonality' column
vc_data['Seasonality'] = vc_data['Incident_From_Date'].dt.month.apply(bin_season)
```

    Visualize/Correct column dtypes

```python
print(vc_data.dtypes)  # Check the data types of your columns

# Columns we want to change from objects into numerical data types
cols_to_convert = ['Census_Tract_Population', 'Census_Tract_Poverty_Population', 'Census_Tract_Poverty_Rate',
                   'Shape_Area', 'Shape_Length', 'Unemployment_Rate', 'Median_Household_Income', 'Mean_Household_Income',
                   'Vehicle_Year']

# Converts the list of columns into int or float64
for col in cols_to_convert:
    vc_data[col] = pd.to_numeric(vc_data[col], errors='coerce')
```

```python
# Group the data by 'NAME10' (census tract) and 'Place2' (private/non-private), and count the crimes
crime_counts = vc_data.groupby(['NAME10', 'Place2']).size().unstack(fill_value=0)

# Find the census tract with the highest count of non-private crimes (Place2 == 0)
highest_non_private = crime_counts[0].idxmax()

# Find the census tract with the highest count of private crimes (Place2 == 1)
highest_private = crime_counts[1].idxmax()

# Filter the dataset for the census tracts of interest
census_tracts_of_interest = vc_data[vc_data['NAME10'].isin([highest_non_private, highest_private])]

# Group the data by 'NAME10' and 'Vehicle_Color', then count the occurrences
color_counts = census_tracts_of_interest.groupby(['NAME10', 'Vehicle_Color']).size().unstack(fill_value=0)

# Select the top 5 vehicle colors for each census tract
top_colors_tract_1 = color_counts.loc[highest_non_private].nlargest(5)
top_colors_tract_43_05 = color_counts.loc[highest_private].nlargest(5)

# Now we will plot only the top 5 colors for each tract. To do this, we create a new DataFrame
top_colors_df = pd.DataFrame({
    f'Census Tract {highest_non_private}': top_colors_tract_1,
    f'Census Tract {highest_private}': top_colors_tract_43_05
}).fillna(0)  # Fill NaN with 0 for colors not present in both tracts

# Plotting the top vehicle colors distribution as a bar chart
ax = top_colors_df.plot(kind='bar', figsize=(10, 6))
plt.title('Distribution of Vehicle Color in Census Tracts with Highest Crime Counts')
plt.ylabel('Number of Vehicles')
plt.xlabel('Vehicle Color')

# Manually setting the x-ticks to the middle of the bars
ax.set_xticks(range(len(top_colors_df)))
ax.set_xticklabels(top_colors_df.index, rotation=0, ha='center')

plt.legend(title='Census Tract')
plt.tight_layout()
plt.show()
```

The distribution of vehicle color across the previously mentioned census tracts.

```python
# Creating a confusion matrix

def make_confusion_matrix(cf,
                          group_names=None,
                          categories='auto',
                          count=True,
                          percent=True,
                          cbar=True,
                          xyticks=True,
                          xyplotlabels=True,
                          sum_stats=True,
                          figsize=None,
                          cmap='Blues',
                          title=None):
    blanks = ['' for i in range(cf.size)]

    if group_names and len(group_names)==cf.size:
        group_labels = ["{}\n".format(value) for value in group_names]
    else:
        group_labels = blanks

    if count:
        group_counts = ["{0:0.0f}\n".format(value) for value in cf.flatten()]
    else:
        group_counts = blanks

    if percent:
        group_percentages = ["{0:.2%}".format(value) for value in cf.flatten()/np.sum(cf)]
    else:
        group_percentages = blanks

    box_labels = [f"{v1}{v2}{v3}".strip() for v1, v2, v3 in zip(group_labels,group_counts,group_percentages)]
    box_labels = np.asarray(box_labels).reshape(cf.shape[0],cf.shape[1])


    # CODE TO GENERATE SUMMARY STATISTICS & TEXT FOR SUMMARY STATS
    if sum_stats:
        #Accuracy is sum of diagonal divided by total observations
        accuracy  = np.trace(cf) / float(np.sum(cf))

        #if it is a binary confusion matrix, show some more stats
        if len(cf)==2:
            #Metrics for Binary Confusion Matrices
            precision = cf[1,1] / sum(cf[:,1])
            recall    = cf[1,1] / sum(cf[1,:])
            f1_score  = 2*precision*recall / (precision + recall)
            stats_text = "\n\nAccuracy={:0.3f}\nPrecision={:0.3f}\nRecall={:0.3f}\nF1 Score={:0.3f}".format(
                accuracy,precision,recall,f1_score)
        else:
            stats_text = "\n\nAccuracy={:0.3f}".format(accuracy)
    else:
        stats_text = ""


    # SET FIGURE PARAMETERS ACCORDING TO OTHER ARGUMENTS
    if figsize==None:
        #Get default figure size if not set
        figsize = plt.rcParams.get('figure.figsize')

    if xyticks==False:
        #Do not show categories if xyticks is False
        categories=False


    # MAKE THE HEATMAP VISUALIZATION
    plt.figure(figsize=figsize)
    sns.heatmap(cf,annot=box_labels,fmt="",cmap=cmap,cbar=cbar,xticklabels=categories,yticklabels=categories)

    if xyplotlabels:
        plt.ylabel('True Place2')
        plt.xlabel('Predicted Place2' + stats_text)
    else:
        plt.xlabel(stats_text)

    if title:
        plt.title(title)
```
